```python
def travel(d,end,cur,costs,passes,tdays):
    if d > end:
        print(cur)
        return 
    if d not in tdays: travel(d+1,end,cur,costs,passes,tdays)
        
    for c in range(len(costs)):
        travel(d+passes[c],end,cur+costs[c],costs,passes,tdays)
            


def mincostTickets(self, days: 'List[int]', costs: 'List[int]') -> 'int':
    passes=[1,7,30]
    cur=0
    end=days[-1]
    travel(0,end,cur,costs,passes,[1,3])

            
        
```
