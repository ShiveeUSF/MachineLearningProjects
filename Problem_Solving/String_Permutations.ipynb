```python
a='ABCDEF'
```

```python
def permute(sfirst,slast,perms):
    if len(slast)==0:
        perms.add(sfirst)
        return sfirst
    for i in range(len(slast)):
        permute(sfirst+slast[i],slast[:i]+slast[i+1:],perms)
```

```python
perms=set()
permute('',a,perms)
```
